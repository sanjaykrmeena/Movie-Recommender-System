```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
credits
```

```python
movies=movies[['movie_id','title','overview','genres','cast','crew','keywords']]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']
```

```python
import ast
def con(obj):
    ans=[]
    for i in ast.literal_eval(obj):
        ans.append(i['name'])
       
      
        
        
        
    return ans
```

```python
movies['genres']=movies['genres'].apply(con)
```

```python
movies['keywords']=movies['keywords'].apply(con)
```

```python
movies.head()
```

```python
def castselector(obj):
    counter=0
    ans=[]
    for i in ast.literal_eval(obj):
        if counter<=3:
            ans.append(i['name'])
            counter+=1
        else:
            break
    return ans
            
```

```python
movies['cast']=movies['cast'].apply(castselector)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
    try:
         text=str(text)
         L = []
         for i in ast.literal_eval(text):
                if i['job'] == 'Director':
                    L.append(i['name'])
         return L 
    except:
#         pass
        return ['NOT KNOWN']
        
   
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
# movies['crew'][0]
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])


```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Batman')
```
